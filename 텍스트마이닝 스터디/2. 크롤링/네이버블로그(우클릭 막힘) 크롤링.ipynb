```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://blog.naver.com/PostView.nhn?blogId=mkt0716&logNo=30031141517&parentCategoryNo=&categoryNo=&viewDate=&isShowPopularPosts=false&from=postView#'
```

```python
source_code = requests.get(url)
```

```python
plain_text = source_code.text
```

```python
plain_text[:1000]
```

```python
soup = BeautifulSoup(plain_text, 'lxml')
```

```python
soup_div = soup.find_all('div',{'id':'post-view30031141517'})
```

```python
for line in soup_div:
    print(line.text)
```
