```python
# 3. twitter API 생성  
api = tweepy.API(auth)

keyword = "게임"     # 자신이 검색하고 싶은 키워드 입력 
search = [] # 크롤링 결과 저장할 변수   

cnt = 1
while(cnt <= 30):   # 10page 대상으로 크롤링
    tweets = api.search(keyword)
    for tweet in tweets:
        search.append(tweet)
    cnt += 1

print(len(search)) # 문서 길이 
print(search[1]) # 첫번째 text 보기 [출처] #파이썬. Python으로 트위터 크롤링(twitter crawling)하기|작성자 파이떤
```

```python
data = {}   # 전체 문서 추가
i = 0       # 문서 번호
for tweet in search:
    data['text'] = tweet.text   # text키에 text문서 저장
    print(i, " : ", data)   # 문서번호 : 문서내용
    i += 1 #파이썬. Python으로 트위터 크롤링(twitter crawling)하기|작성자 파이떤
```

```python
class listener(tweepy.streaming.StreamListener):
    def on_data(self, data):
        try:
            data = data.encode('utf-8')
            data = data.decode('unicode_escape')
        except:
            print('Err')

        print(data)
        return True
    def on_error(self, status):
        print('error : ', status)
```

```python
twitterStream = tweepy.Stream(auth, listener())
twitterStream.filter(track='게임')
```
